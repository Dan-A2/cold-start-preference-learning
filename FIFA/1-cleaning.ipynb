# Cleaning the dataset

Firstly, I noticed that some of the columns are urls. We don't need them and we will delete them.

```python
url_columns = [col for col in data.columns if 'url' in col]
print(url_columns)
df_cleaned = data.drop(columns=url_columns)
```

```python
nan_values = df_cleaned.isna().sum()
print(nan_values[nan_values > 0])
```

```python
df_cleaned[nan_values[nan_values > 0].index].head()
```

```python
df_cleaned[df_cleaned['club_loaned_from'].isna() == False]['club_loaned_from'].head()
```

As we can see, some of the values of very important columns are missing, but they are a small proportion of our dataset so we will delete them. These include the columns with less than 100 missing values.

```python
df_cleaned.dropna(subset=list(nan_values[(nan_values > 0) & (nan_values < 100)].index), inplace=True)
nan_values = df_cleaned.isna().sum()
print(nan_values[nan_values > 0])
```

```python
print(len(df_cleaned))
```

The goal keeping speed is the speed rating for goal keepers. These values must be Nan for non-goalers which is completely natural. In order to decide about what to do with this column, we wil try to find out the influence of tis column to the value of goal keepers.

```python
df_cleaned['goalkeeping_speed'].value_counts().index
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned[df_cleaned['goalkeeping_speed'].isna() == False], x='goalkeeping_speed', y='value_eur')
plt.title('Correlation between Goalkeeping Speed and Player Value')
plt.xlabel('Goalkeeping Speed')
plt.ylabel('Player Value (EUR)')
plt.show()
```

As we can see, the correlation can't be ignored. In order to fix the issue of nan values for this column, we will put 0 as the goal keeping speed of non goal keepers.

```python
df_cleaned['goalkeeping_speed'] = df_cleaned['goalkeeping_speed'].fillna(0)
```

```python
nan_values = df_cleaned.isna().sum()
print(nan_values[nan_values > 0])
```

Now we will drop all the other columns with more than 17000 missing values because more than 95% of those columns is nan and there is no sensible way to fill them.

```python
df_cleaned.drop(columns=list(nan_values[nan_values > 17000].index), inplace=True)
nan_values = df_cleaned.isna().sum()
print(nan_values[nan_values > 0])
```

Let's take a closer look to these columns and then we will decide what to do about them.

```python
df_cleaned[nan_values[nan_values > 0].index].head()
```

We know that the joining date is completely unrelated to the value of a player (or has very low correlation) so we will justt delete this column. Also the player_traits column is in the string format and is not useful for our analysis so we will also remove that one.

```python
df_cleaned.drop(columns=['club_joined', 'player_traits'], inplace=True)
```

Now let's consider other columns

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='release_clause_eur', y='value_eur')
plt.title('Correlation between Release Clause and Player Value')
plt.xlabel('Release Clause')
plt.ylabel('Player Value (EUR)')
plt.show()
```

As we can see, this is an extremely important feature because it has a great linear correlation with the player value. So we will drop the rows which don't have this value.

```python
df_cleaned.dropna(subset=['release_clause_eur'], inplace=True)
nan_values = df_cleaned.isna().sum()
print(nan_values[nan_values > 0])
print(len(df_cleaned))
```

```python
df_cleaned_gk = df_cleaned[df_cleaned['club_position'] == 'GK']
nan_values = df_cleaned_gk.isna().sum()
print(nan_values[nan_values > 0])
print(len(df_cleaned_gk))
```

As we can see, all the goal keepers have these values missing. So we will just replace it with 0.

```python
df_cleaned.loc[df_cleaned['club_position'] == 'GK'] = (
    df_cleaned.loc[df_cleaned['club_position'] == 'GK'].fillna(0)
)
```

Now we will plot all of these values and check the correlation with player value.

```python
columns_to_plot = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Correlation of Player Attributes with Player Value', fontsize=16)

titles = ['Pace', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physic']

for ax, col, title in zip(axes.flat, columns_to_plot, titles):
    sns.regplot(data=df_cleaned[df_cleaned['club_position'] != 'GK'], x=col, y='value_eur', order=2, ax=ax, line_kws={'color': 'red'})
    ax.set_title(f'{title} vs Player Value')
    ax.set_xlabel(title)
    ax.set_ylabel('Player Value (EUR)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

There is little nonlinear relationship between them. Because the dataset is huge we can drop the Nan values of these columns and use them.

```python
df_cleaned.dropna(subset=nan_values.index, inplace=True)
```

Now let's check the columns the datatype of which is object.

```python
object_columns = df_cleaned.select_dtypes(include=['object']).columns.tolist()
print(object_columns)
```

```python
df_cleaned[object_columns].head()
```

Ofcourse the name of the player has little to do with his value (except for the most famous ones which can be ignored) so we will delete the long name column. We will keep the short_name column and rename it to name so that we know who we are talking about (not used for training!). We will also delete the dob(date of birth) column and replace it with age. Afterwards we will decide about what to do with the rest of the columns.

```python
df_cleaned.drop(columns=['long_name'], inplace=True)
df_cleaned.rename(columns={'short_name': 'name'}, inplace=True)

df_cleaned['dob'] = pd.to_datetime(df_cleaned['dob'])
current_year = pd.to_datetime('now').year
df_cleaned['age'] = current_year - df_cleaned['dob'].dt.year
df_cleaned.drop(columns=['dob'], inplace=True)
```

For the last columns in the list, which are the skill measure of the player in that role, we can see that the values are something like 80+3. The 80 is the basic amount of this attribute ant the +3 part is tha amount that in can be increased in the game based on different in-game stuff. So for this analysis we will just put the basic integer value for these columns.

```python
rating_columns = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram','lm', 'lcm', 'cm',
                  'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']

for col in rating_columns:
    df_cleaned[col] = df_cleaned[col].apply(lambda x: int(x.split('+')[0]) if '+' in x else int(x.split('-')[0]))
```

Now we will check the correlation between each one of the object columns with our target value.

```python
categorical_columns = ['player_positions', 'club_name', 'league_name', 'club_position', 'nationality_name', 
                       'work_rate', 'body_type', 'real_face', 'preferred_foot']

fig, axes = plt.subplots(9, 1, figsize=(50, 80))

for i, col in enumerate(categorical_columns):
    # Convert the column to numerical representation
    df_encoded = pd.get_dummies(df_cleaned[col], drop_first=True)
    df_encoded['value_eur'] = df_cleaned['value_eur']
    correlation = df_encoded.corr()['value_eur']
    
    # Plot correlation for each feature
    correlation.drop('value_eur').plot(kind='bar', ax=axes[i], color='skyblue')
    axes[i].set_title(col)
    axes[i].set_xlabel('Feature')
    axes[i].set_ylabel('Correlation Coefficient')

# Adjust layout
plt.tight_layout()
plt.show()
```

As we can see all of them except preferred foot are correlated with our target value. The correlation coefficient for the preffered fott column is small (around -0.01), but we will keep it because it may be usefull via correlation with other columns.

After looking to dataset columns, some of them seem to be useless for our prediction. We will check if they are actually useless or not. We will start with age, height and weight.

```python
sns.set_style("whitegrid")
g = sns.PairGrid(df_cleaned, x_vars=["age", "height_cm", "weight_kg"], y_vars=["value_eur"], height=4, aspect=1.5)
g.map(sns.regplot, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.tight_layout()
plt.show()
```

As we can see there is no specific relationship between these and our target value. So we will delete all these columns.

```python
df_cleaned.drop(columns=["age", "height_cm", "weight_kg"], inplace=True)
```

There are two columns named club_team_id and club_name. They basically do the same thing so we will delete the club_name column. Same happens for nationality_id and nationality_name.

```python
df_cleaned.drop(columns=["club_name", "nationality_name"], inplace=True)
```
